```python
import random

class SlayweksChatbot:
    def __init__(self):
        self.welcome_responses = ["Welcome to Slayweks! How can I assist you today?",
                                  "Hello! Welcome to Slayweks. What service are you looking for?",
                                  "Hi there! Glad to see you on Slayweks. How can I help you?"]

        self.locations = ["Anna Nagar", "Tambaram", "Guindy", "Chennai Central", "Sholinganallur"]

        self.services = ["Haircut", "Manicure", "Pedicure", "Salon", "Facial", "Nails"]

        # In a real scenario, you would have a database to store salon details
        self.salons = {
            "Anna Nagar": {
                "Salon1": {
                    "Services": ["Haircut", "Manicure", "Pedicure"],
                    "Contact": "123-456-7890"
                },
                "Salon2": {
                    "Services": ["Haircut", "Facial", "Nails"],
                    "Contact": "234-567-8901"
                }
            },
            "Tambaram": {
                "Salon3": {
                    "Services": ["Haircut", "Manicure", "Pedicure"],
                    "Contact": "345-678-9012"
                },
                "Salon4": {
                    "Services": ["Haircut", "Salon", "Facial"],
                    "Contact": "456-789-0123"
                }
            },
            "Guindy": {
                "Salon5": {
                    "Services": ["Haircut", "Manicure", "Pedicure"],
                    "Contact": "567-890-1234"
                },
                "Salon6": {
                    "Services": ["Haircut", "Facial", "Nails"],
                    "Contact": "678-901-2345"
                }
            },
            "Chennai Central": {
                "Salon7": {
                    "Services": ["Haircut", "Manicure", "Pedicure"],
                    "Contact": "789-012-3456"
                },
                "Salon8": {
                    "Services": ["Haircut", "Salon", "Facial"],
                    "Contact": "890-123-4567"
                }
            },
            "Sholinganallur": {
                "Salon9": {
                    "Services": ["Haircut", "Manicure", "Pedicure"],
                    "Contact": "901-234-5678"
                },
                "Salon10": {
                    "Services": ["Haircut", "Facial", "Nails"],
                    "Contact": "012-345-6789"
                }
            }
        }

    def welcome_user(self):
        return random.choice(self.welcome_responses)

    def get_location_options(self):
        return self.locations

    def get_service_options(self):
        return self.services

    def get_salon_options(self, location, service):
        if location in self.salons and service:
            return list(self.salons[location].keys())
        else:
            return []

    def provide_contact_details(self, location, selected_salon):
        if location in self.salons and selected_salon in self.salons[location]:
            return self.salons[location][selected_salon]["Contact"]
        else:
            return "Contact details not available"

    def chat(self):
        print(self.welcome_user())

        print("Available Locations:")
        for index, location in enumerate(self.get_location_options(), start=1):
            print(f"{index}. {location}")

        location_choice = int(input("Please select a location number: "))
        location = self.locations[location_choice - 1]

        print("Available Services:")
        for index, service in enumerate(self.get_service_options(), start=1):
            print(f"{index}. {service}")

        service_choice = int(input("Please select a service number: "))
        service = self.services[service_choice - 1]

        salon_options = self.get_salon_options(location, service)

        if salon_options:
            print(f"Here are some salon options near {location} that provide {service}:")
            for index, salon in enumerate(salon_options, start=1):
                print(f"{index}. {salon}")

            choice = input("Please select a salon number: ")
            selected_salon = salon_options[int(choice) - 1]
            contact_details = self.provide_contact_details(location, selected_salon)
            print(f"Contact details for {selected_salon}: {contact_details}")
        else:
            print("No salon found for the given location and service")

if __name__ == "__main__":
    chatbot = SlayweksChatbot()
    chatbot.chat()
```
